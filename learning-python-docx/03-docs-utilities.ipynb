```python
from docx import Document
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import Pt
```

### Alignment / Justify

```python
doc = Document()

p = doc.add_paragraph("This is a paragraph will be centered")
p.alignment = WD_ALIGN_PARAGRAPH.CENTER

p = doc.add_paragraph("This is a paragraph will be right")
p.alignment = WD_ALIGN_PARAGRAPH.RIGHT

p = doc.add_paragraph("This is a paragraph will be left")
p.alignment = WD_ALIGN_PARAGRAPH.LEFT

p = doc.add_paragraph("This is a paragraph will be justify")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

doc.save("../results/aligned.docx")
```

### Indentation

```python
p = doc.add_paragraph("This paragraph has indentation.")

# First-line indent
p.paragraph_format.first_line_indent = Pt(24) 

# # Left / Right indentation
p.paragraph_format.left_indent = Pt(36)

p = doc.add_paragraph("This is spacing")
p.paragraph_format.space_before = Pt(12)

p = doc.add_paragraph("This is spacing")
p.paragraph_format.space_after = Pt(12)

from docx.shared import Inches

# TTABS
p = doc.add_paragraph("Text with tab")
p.paragraph_format.tab_stops.add_tab_stop(Inches(1.5))

doc.save("../results/aligned.docx")
```
