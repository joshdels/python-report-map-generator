Exercise 1 — Multiple Paragraphs & Text Formatting 
Goal: Add multiple paragraphs with basic formatting.

```python
document = Document()

document.add_heading("Practice 1")
parangraph = document.add_paragraph("Hello Joshua ")
bold_run = parangraph.add_run("Gwapo")
bold_run.bold = True

parangraph = document.add_paragraph("It is I who conquers the world ")
underline = parangraph.add_run("Philippines")
underline.underline = True

document.save("results/exercise1.docx")
```

Exercise 2 — Headings & Structure  
Goal: Structure a document using headings.

```python
document = Document()

document.add_heading("Practice 2", level=1)
document.add_heading("This is sublevel", level=2)

parangraph = document.add_paragraph("This is me geospatial developer")
document.add_heading("features", level=2)
document.add_paragraph("list", style="List")
document.add_paragraph("tables", style="List Bullet")
document.add_paragraph("new line", style="List Bullet 2")

document.save("results/exercise2.docx")
```

Exercise 4 — Lists (Bullet & Numbered)
Goal: Create bullet and numbered lists.

```python
document = Document()

document.add_heading("Shopping List")

lists = ["Rice", "Milk", "Eggs"]
routines = ["Wakeup", "Code", "Sleep"]

def list_holder(lists_array, style):
  for i in lists_array:
      parangraph = document.add_paragraph(i, style=style)

list_holder(lists, "List Bullet")

document.add_heading("Daily Routine")
list_holder(routines, "List Number")

document.save("results/exercise4.docx")
```

Exercise 5 — Tables (Core Fundamental)
Goal: Create and populate a table.

```python
document = Document()

table = document.add_table(rows=3, cols=3)

cell = table.cell(0,0)
cell.text = 'Name'
cell = table.cell(0,1)
cell.text = 'Age'
cell = table.cell(0,2)
cell.text = 'Country'

document.save("results/exercise5.docx")
```

Exercise 6 — Reading Data & Writing to Word
Goal: Use Python data structures with Word.

```python
document = Document()

table = document.add_table(rows=1, cols=3)

# Table Construction
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Name'
hdr_cells[1].text = 'Age'
hdr_cells[2].text = 'Country'

students = [
    ("Joshua", 21, "Philippines"),
    ("Ana", 22, "Japan"),
    ("Mark", 20, "USA"),
    ("Mark", 20, "USA"),
    ("Mark", 30, "USA")
]

for student in students:
  row_cells = table.add_row().cells
  for col_index, info in enumerate(student):
    row_cells[col_index].text = str(info)
    

document.save("../results/exercise6.docx")
```

Exercise 7 — Mini Project (Fundamentals Challenge)
Goal: Generate a simple report.

```python
from datetime import datetime
date = datetime.now().strftime("%B %d, %Y")

document = Document()

document.add_heading('Weekly Progress Report')
document.add_heading('Date: ' + date , level=3)

tasks = [
  ('running', 5, "well done"),
  ('swimming', 2, "nicely done"),
  ('cooking', 1, "failed"),
]

# TASK COMPLETED
document.add_paragraph("")
section = document.add_heading("Section: Tasks Completed", level=2)

for task in tasks:
  document.add_paragraph(task[0], style="List Bullet")

# HOURS SPENT
document.add_paragraph("")
section = document.add_heading("Section: Hours Spent", level=2)

table = document.add_table(rows=1, cols=2)

hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Task'
hdr_cells[1].text = 'Hours'

for task in tasks:
  row_cells = table.add_row().cells
  row_cells[0].text = task[0]
  row_cells[1].text = str(task[1])

# NOTES
document.add_paragraph("")
section = document.add_heading("Section: Notes", level=2)

for task in tasks:
  document.add_paragraph(f"{task[0]}: {task[2]}")

document.save("../results/weekly_report.docx")
```

```python
# TO PDF
import subprocess

output_folder = "../results/"

subprocess.run([
    "libreoffice",
    "--headless",
    "--convert-to",
    "pdf",
    "../results/weekly_report.docx"
], cwd=output_folder)
```
