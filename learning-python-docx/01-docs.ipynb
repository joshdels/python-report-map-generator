### Learning the basic of python docs for automated reports

More in this documentation: https://python-docx.readthedocs.io/en/latest/user/quickstart.html

```python
from docx import Document
document = Document()

# document.save('results/save.docx')
```

### Writing Paragraphs

```python
document = Document()

paragraph = document.add_paragraph('Lorem 21342')
prior_paragraph = paragraph.insert_paragraph_before("2nd paragraph")

document.add_heading('The heading')
document.add_heading('The heading 2')

document.add_heading('The level 2', level=2)
document.add_heading('The level 3', level=3)

document.add_heading('Document Title', 0)

# new page
document.add_page_break()

document.save('../results/demo1.docx')
```

#### Adding Table

```python
document = Document()

table = document.add_table(rows=2, cols=3)
cell = table.cell(0,0)
cell.text = 'parrot, possibly dead'

document.add_page_break()

table = document.add_table(rows=1, cols=2)
cell = table.cell(0,1)
cell.text = 'hey testing'

document.save('results/tables.docx')
```

#### Adding Picture

```python
from docx.shared import Inches

for i in range(5):
  document.add_picture('results/image.jpg', width=Inches(1.0))

document.save('results/pictures.docx')
```

### Adding Styles

```python

document = Document()

# bold
paragraph = document.add_paragraph("Lorem this is paragaraph", style='ListBullet')
bold_run = paragraph.add_run(' boldy my goodness')
bold_run.bold = True

# italic
itallic_run = paragraph.add_run( 'italic baby')
itallic_run.italic = True

document.save('results/styles.docx')
```

### Adding Colors

```python
from docx.shared import RGBColor

document = Document()

paragraph = document.add_paragraph()
run = paragraph.add_run("Heloo red")
run.font.color.rgb = RGBColor(255, 0, 0)


document.save('../results/colors.docx')
```

### Font Styles

```python
document = Document()

paragraph = document.add_paragraph()
run = paragraph.add_run("This is a new style font")
run.font.name = "Calibre"

document.save('../results/font.docx')
```

### Font Size

```python
from docx.shared import Pt

doc = Document()
p = doc.add_paragraph()
run = p.add_run("This is a new style font")
run.font.size = Pt(50)

doc.save('../results/size.docx')
```

### Indentation

```python
from docx import Document
from docx.enum.text import WD_ALIGN_PARAGRAPH

doc = Document()

p = doc.add_paragraph("This is a paragraph will be centered")
p.alignment = WD_ALIGN_PARAGRAPH.CENTER

p = doc.add_paragraph("This is a paragraph will be right")
p.alignment = WD_ALIGN_PARAGRAPH.RIGHT

p = doc.add_paragraph("This is a paragraph will be left")
p.alignment = WD_ALIGN_PARAGRAPH.LEFT

p = doc.add_paragraph("This is a paragraph will be justify")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

doc.save("../results/aligned.docx")
```
