```python
from docx import Document
from docx.shared import RGBColor

document = Document()

heading = document.add_heading('Contract of Lease', level=1)
for run in heading.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)

subheading = document.add_heading('KNOW ALL BY MANS BY THESES PRESENTS', level=2)
for run in subheading.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)

paragraph = document.add_paragraph("This ")
bold_run = paragraph.add_run("CONTRACT OF LEASES")
bold_run.bold = True
paragraph.add_run(" made and entered into by and between:")

paragraph = document.add_paragraph("")
lessor_name = "Elvie S. Damang"
paragraph = document.add_paragraph()
bold_name = paragraph.add_run(lessor_name.upper())
bold_name.bold = True
paragraph.add_run(", of legal age, married and a resident of Brgy. New Pandan, Panabo City, Philippines, now and heren after called the")
bold_lessor = paragraph.add_run(" LESSOR; ")
bold_lessor.bold = True
paragraph.add_run("  -and--")

lessor_name = "Joshua S. De Leon"
paragraph = document.add_paragraph()
bold_name = paragraph.add_run(lessor_name.upper())
bold_name.bold = True
paragraph.add_run(", also a Filipino, of legal age, single and a resident of 10 Narra St. J. Village, Visayan Village, Tagum City, Davao del Norte, Philippines, now and hereinafter known as the ")
bold_lessor = paragraph.add_run(" LESSEE; ")
bold_lessor.bold = True


# Save file
document.save("../results/letter.docx")
print("✅ File saved!")
```
