```python
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import Pt

document = Document()

# CONFIG ---------------------------------------

LESSOR = {
    "name": "MONICA S. DELA CRUZ",
    "address": "Brgy. New Pandan, Davao City, Philippines",
    "id": "SSD ID No.0X-123123123-X"
}

LESSEE = {
    "name": "JOSHUA S. DE LEON",
    "address": "Brgy. Visayan Village, Tagum City, Philippines",
    "id": "PRC REG. NO.X00232322X"
}

PROPERTY = {
    "name": "2 Doors Apartment",
    "location": "Purok 7, Dead End of Pelpinosas Subd,",
    "lot_num": "500-G-32-A-C",
    "license": "TCT No. 100-X01505000"
    
}

RENT = {
    "cost_num": 6000,
    "cost_string": "six thousand",
    "start_date": "October 4, 2025"
}

HOUSE_RULES = [
    "That the LESSOR shall not be liable for any damage by caused by or any damage arising from the failure of the water supply and or electricity installment or the bursting leaking or running of any wash pipes, rain water or water closet and the the LESSOR shall not be liable",
    "No Pets Allowed inside the appartment",
    "The LESSEE shall not sublease the apartment without the written consent of the LESSOR;",
    "No illegal activities shall be conducted within the premises"
]   

INDENT = Pt(36)

#  ---------------------------------------------------------

heading = document.add_heading('CONTRACT OF LEASE')
heading.bold = True
heading.alignment = WD_ALIGN_PARAGRAPH.CENTER
for run in heading.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)

subheading = document.add_heading('KNOW ALL BY MANS BY THESES PRESENTS', level=2)
for run in subheading.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)

p = document.add_paragraph("This ")
b = p.add_run("CONTRACT OF LEASES")
b.bold = True
p.add_run(" made and entered into by and between:")
p.first_line_indent = Pt(24)
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph()
bold_name = p.add_run(LESSOR["name"])
bold_name.bold = True
p.add_run(", of legal age, married and a resident of  ")
b = p.add_run(LESSOR["address"])
p.add_run(" now and heren after called the")
b = p.add_run(" LESSOR; ")
b.bold = True
p.add_run("  -and--")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph()
bold_name = p.add_run(LESSEE["name"])
bold_name.bold = True
p.add_run(", also a Filipino, of legal age, single and a resident of ")
b = p.add_run(LESSEE["address"])
p.add_run(" now and hereinafter known as the ")
b = p.add_run(" LESSEE; ")
b.bold = True
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

withnesseth = document.add_heading("WITHNESSETH:")
withnesseth.alignment = WD_ALIGN_PARAGRAPH.CENTER
for run in withnesseth.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)

# MAIN BODY
p = document.add_paragraph("That herein the LESSOR ")
b = p.add_run(LESSOR["name"])
b.bold = True
p.add_run(" is known as the ")
b = p.add_run("OWNER ")
b.bold = True
p.add_run(f" of {PROPERTY['name']} at {PROPERTY['location']}")
p.add_run(f" covered by {PROPERTY['license']}.")
p.paragraph_format.first_line_indent = INDENT
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph("That the both parties have agreed to lease the Door one (1) and let said above property subject to the following terms and conditions:")
p.paragraph_format.first_line_indent = INDENT
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph("That the term of this lease contract shall be for the period of")
p.add_run(" year, to commence ")
b = p.add_run(RENT['start_date'])
b.bold = True
p.add_run(" with monthly rental of ")
b = p.add_run(f"{RENT['cost_string'].upper()} Php {RENT['cost_num']} PESOS")
b.bold = True
p.add_run(", renewable to another term at the option of the ")
b = p.add_run("LESSOR;")
b.bold = True
p.paragraph_format.first_line_indent = INDENT
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

# HOUSE RULES
subheading = document.add_heading('HOUSE RULES:', level=2)
for run in subheading.runs:
    run.font.color.rgb = RGBColor(0, 0, 0)
    
for i in HOUSE_RULES:
    p = document.add_paragraph(i, "List Bullet")
    p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

# AGREEMENT
p = document.add_paragraph("That either parties may teminate the lease by giving at least seven (7) days written notice. The ")
b = p.add_run("LESSE ")
b.bold = True
p.add_run("must vacate the premises in good and clean condition.")    
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph("IN WITNESS WHEREOF, the parties hereto set their hands this __________, at ___________, Philippines.")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

# TABLE
p = document.add_paragraph()
b = p.add_run(LESSOR['name'])
b.bold = True
p = document.add_paragraph("Lessor")
p = document.add_paragraph(LESSOR['id'])

p = document.add_paragraph()
b = p.add_run(LESSEE['name'])
b.bold = True
p = document.add_paragraph("Lessee")
p = document.add_paragraph(LESSEE['id'])

p = document.add_paragraph(" Signed in the pressence of ___________ and _____________")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

p = document.add_paragraph(
    "BEFORE ME, a Notary Public for and the city of Davao/Province of Davao, personally appered the above name with their valid ID's written below their names, ")
p.add_run("which known to me to the very same persons who executed the foregoing instrument andthey acknowledge to me the same is their free and voluntary deed.")
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
p.paragraph_format.first_line_indent = Pt(36)


# SAVE FILE
document.save("../results/letter.docx")
print("✅ File saved!")
```
